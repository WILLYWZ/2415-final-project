```python
import csv
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import files

# data from https://data.oecd.org/conversion/exchange-rates.htm
data = files.upload()
```

```python
x = []
y = []
y1 = []
y2 = []
y3 = []
y4 = []

with open('dataGraph.csv') as file:
  plots = csv.reader(file, delimiter = ',')

  for row in plots:

    if row[0]=="UK GBP":
      x.append(row[1])
      y.append(float(row[2])) 
    elif row[0]=="European Euro":
      y1.append(float(row[2]))
    elif row[0]=="AUS AUD":
      y2.append(float(row[2]))
    elif row[0]=="New Zealand NZ":
      y3.append(float(row[2]))
    elif row[0]=="US USD":
      y4.append(float(row[2]))

fig = plt.figure(figsize=(16,6))
ax = fig.add_subplot(111)

for i,j in zip(x,y):
    ax.annotate(str(j),xy=(i,j-0.1),color='forestgreen')

for i,j in zip(x,y1):
    ax.annotate(str(j),xy=(i,j+0.03),color='orangered')

for i,j in zip(x,y2):
    ax.annotate(str(j),xy=(i,j-0.08),color='royalblue')
  
for i,j in zip(x,y3):
    ax.annotate(str(j),xy=(i,j+0.04),color='purple')


plt.plot(x , y, marker = 'o', color = 'forestgreen', label="UK GBP")
plt.plot(x , y1, marker = 'o', color = 'orangered', label="European Euro")
plt.plot(x , y2, marker = 'o', color = 'royalblue', label="AUS AUD")
plt.plot(x , y3, marker = 'o', color = 'purple', label="New Zealand NZ")
plt.plot(x , y4, marker = 'o', color = 'deeppink', label="US USD")


plt.ylim([0.25,2.5])
plt.title('Major currency exchange rate in USD from 2000 to 2021')
plt.ylabel('Exchange Rate')
plt.xlabel('Years')

leg = plt.legend(loc='upper right')

plt.show()


```

```python
fig = plt.figure(figsize=(16,6))
ax = fig.add_subplot(111)

for i,j in zip(x,y):
    ax.annotate(str(j),xy=(i,j-0.1),color='blue')

for i,j in zip(x,y1):
    ax.annotate(str(j),xy=(i,j+0.03),color='red')

for i,j in zip(x,y2):
    ax.annotate(str(j),xy=(i,j-0.02),color='green')
  
for i,j in zip(x,y3):
    ax.annotate(str(j),xy=(i,j+0.04),color='darkorange')

a = plt.stem(x,y,linefmt='royalblue', markerfmt='*', bottom=1, label="UK GBP", use_line_collection=True)
b = plt.stem(x,y3,linefmt='darkorange', markerfmt='^', bottom=1,label="New Zealand NZ", use_line_collection=True)
c = plt.stem(x,y2,linefmt='green', markerfmt='v', bottom=1, label="AUS AUD", use_line_collection=True)
d = plt.stem(x,y1,linefmt='red', markerfmt='p', bottom=1,label="European Euro", use_line_collection=True)

plt.legend(loc='upper right')

plt.show()
```

```python
df = pd.DataFrame(dict(graph=x,
                           p=y, p1 =y1, p2 =y2, p3=y3))
pad = 1
bar = 0.5
barRange = np.arange(len(x))*(4*bar+pad)

fig = plt.figure(figsize=(20,20))
plt.barh(barRange, df.p,  color='purple', label= "European Euro")
plt.barh(barRange+bar, df.p1, color='lightseagreen', label= "New Zealand NZ")
plt.barh(barRange+2*bar, df.p2, color='palegoldenrod', label= "AUS AUD")
plt.barh(barRange+3*bar, df.p3, color='lightskyblue', label= "European Euro")

ax = fig.add_subplot()


plt.ylabel('Year', rotation = 0, fontweight="bold", size = 20)
ax.yaxis.set_label_coords(-0.07,0.50)

plt.title('Major currency exchange rate in USD from 2000 to 2021', fontweight="bold", size = 25)
plt.xlabel('Exchange Rate', fontweight="bold", size = 20)
plt.yticks(barRange, x)
plt.legend(loc= 'right')


plt.ylim(-0.5, 64.8)
plt.show()
```
